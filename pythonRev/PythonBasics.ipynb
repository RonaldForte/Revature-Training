```python
#Slide Question: Format a string value
# Remove whitespace, convert to all lowercase, replace '-' with '_', remove all '!', check atleast one digit is present, check it contains 'python', create a variable that stores only the first 8 chars
raw_username = " Py_Thon-ROCKS123!!!"
formatted_username = raw_username.strip().lower().replace('-', '_').replace('!', '')
digits = 0
for x in formatted_username:
    if x.isdigit():
        digits += 1
print(formatted_username)
print("There is " + str(digits) + " digits in this string.")
if "python" in formatted_username:
    print("This string contains the word Python.")
first_eight = formatted_username[0:8]
print("The first eight characters are " + first_eight)
#Learn f string no time during lecture
```

```python
#Slide Question: Work with Lambdas
students = [
    ('alice', 34),
    ('bob', 22),
    ('diana', 19)
]

newList = sorted(students, key=lambda student: student[1]) #reverse if backwards

for student in newList:
    print(student)
```
